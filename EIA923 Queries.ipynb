```python
keywords = ['NextEra', 'FPL', 'Florida Power']

# Create an empty list to store the matching names
matching_names = []

for value in df21['Operator Name']:
    # Convert the value to a string if it is a float
    if isinstance(value, float):
        value = str(value)
    
    # Check if the value contains any of the keywords
    if any(keyword in value for keyword in keywords):
        matching_names.append(value)

```

```python
unique_names = list(set(matching_names))
```

```python
unique_names
```

```python
# Filter the original DataFrame based on the unique names
filtered_df = df21[df21['Operator Name'].isin(unique_names)]

# Calculate the sum of net generation for the filtered names
sum_net_generation = filtered_df['Net Generation\n(Megawatthours)'].sum()
```
